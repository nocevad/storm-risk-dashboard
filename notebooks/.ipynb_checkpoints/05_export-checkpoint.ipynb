```python
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import geopandas as gpd
import imageio.v2 as imageio
import os
import requests
import tempfile

load_dotenv(r"C:\Projects\storm_risk\.env")

engine = create_engine(
    "mysql+mysqlconnector://",
    connect_args={
        "host"    : os.getenv('MYSQL_HOST'),
        "user"    : os.getenv('MYSQL_USER'),
        "password": os.getenv('MYSQL_PASSWORD'),
        "database": os.getenv('MYSQL_DATABASE')
    }
)

print("✅ Connected!")
```

```python
# Load county GeoJSON as a GeoDataFrame for matplotlib plotting
geojson_url = "https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json"
counties_geojson = requests.get(geojson_url).json()

# Convert to GeoDataFrame
counties_gdf = gpd.GeoDataFrame.from_features(
    counties_geojson['features']
)

# The FIPS code is stored in the 'id' field
counties_gdf['fips_code'] = counties_gdf['id'].str.zfill(5)

# Filter to continental US only by bounding box
# This removes Alaska, Hawaii, Puerto Rico from the map
counties_gdf = counties_gdf.cx[-125:-65, 24:50]

print(f"Counties in GeoDataFrame: {len(counties_gdf)}")
print(f"Columns: {counties_gdf.columns.tolist()}")
```

```python
# Load county GeoJSON as a GeoDataFrame
geojson_url = "https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json"
counties_geojson = requests.get(geojson_url).json()

# Convert to GeoDataFrame - pass id from each feature manually
counties_gdf = gpd.GeoDataFrame.from_features(
    counties_geojson['features']
)

# Extract FIPS from the features list directly
fips_list = [f['id'] for f in counties_geojson['features']]
counties_gdf['fips_code'] = [str(f).zfill(5) for f in fips_list]

# Check what we have
print(f"Columns: {counties_gdf.columns.tolist()}")
print(f"Total counties: {len(counties_gdf)}")
print(f"\nSample fips codes: {counties_gdf['fips_code'].head().tolist()}")

# Filter to continental US only
counties_gdf = counties_gdf.cx[-125:-65, 24:50]

print(f"\nContinental US counties: {len(counties_gdf)}")
```

```python
# Pull monthly risk from MySQL
# We'll focus on the last 30 years to keep the GIF manageable
monthly_risk = pd.read_sql("""
    SELECT 
        fips_code,
        month,
        enhanced_risk
    FROM monthly_risk
    WHERE month >= '1995-01'
    ORDER BY month ASC
""", engine)

# Get sorted list of unique months for animation frames
months = sorted(monthly_risk['month'].unique())

print(f"Monthly records loaded: {len(monthly_risk)}")
print(f"Animation frames: {len(months)}")
print(f"Date range: {months[0]} to {months[-1]}")
```

```python
# Filter to last 10 years, hurricane season months only (June-October)
hurricane_season_months = ['06','07','08','09','10']

monthly_risk_filtered = monthly_risk[
    (monthly_risk['month'] >= '2015-01') &
    (monthly_risk['month'].str[5:7].isin(hurricane_season_months))
]

# Get sorted unique months
months = sorted(monthly_risk_filtered['month'].unique())

print(f"Filtered records: {len(monthly_risk_filtered)}")
print(f"Animation frames: {len(months)}")
print(f"Date range: {months[0]} to {months[-1]}")
print(f"\nFrames list:")
for m in months:
    print(f"  {m}")
```

```python
# Get global min/max for consistent color scale across all frames
risk_min = monthly_risk_filtered['enhanced_risk'].min()
risk_max = monthly_risk_filtered['enhanced_risk'].max()

print(f"Risk range: {risk_min:.4f} to {risk_max:.4f}")

# Set up color normalization and colormap
norm     = mcolors.Normalize(vmin=risk_min, vmax=risk_max)
colormap = plt.cm.Reds

print(f"✅ Color scale ready")
```

```python
# Test with August 2017 - Hurricane Harvey month
test_month = '2017-08'
month_data = monthly_risk_filtered[
    monthly_risk_filtered['month'] == test_month
]

# Merge risk data onto GeoDataFrame
frame_gdf = counties_gdf.merge(
    month_data[['fips_code','enhanced_risk']],
    on='fips_code',
    how='left'
)
frame_gdf['enhanced_risk'] = frame_gdf['enhanced_risk'].fillna(0)

# Plot
fig, ax = plt.subplots(1, 1, figsize=(14, 8))

frame_gdf.plot(
    column='enhanced_risk',
    ax=ax,
    cmap='Reds',
    norm=norm,
    edgecolor='none',
    linewidth=0
)

ax.set_title(f'U.S. Storm Risk — {test_month}', 
             fontsize=16, fontweight='bold', pad=15)
ax.axis('off')

plt.tight_layout()
plt.savefig(r'C:\Projects\storm_risk\outputs\test_frame.png', 
            dpi=100, bbox_inches='tight')
plt.show()

print("✅ Test frame saved to outputs folder!")
```

```python
# Output path
gif_path = r'C:\Projects\storm_risk\outputs\storm_risk_animation.gif'
temp_frames = []

print("Building frames...")

for i, month in enumerate(months):
    # Get data for this month
    month_data = monthly_risk_filtered[
        monthly_risk_filtered['month'] == month
    ]
    
    # Merge onto GeoDataFrame
    frame_gdf = counties_gdf.merge(
        month_data[['fips_code','enhanced_risk']],
        on='fips_code',
        how='left'
    )
    frame_gdf['enhanced_risk'] = frame_gdf['enhanced_risk'].fillna(0)
    
    # Plot frame
    fig, ax = plt.subplots(1, 1, figsize=(14, 8))
    
    frame_gdf.plot(
        column='enhanced_risk',
        ax=ax,
        cmap='Reds',
        norm=norm,
        edgecolor='none',
        linewidth=0
    )
    
    # Add title with month
    ax.set_title(f'U.S. Storm Risk — {month}',
                 fontsize=16, fontweight='bold', pad=15)
    ax.axis('off')
    
    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap='Reds', norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=ax, orientation='horizontal',
                 fraction=0.03, pad=0.02, label='Enhanced Risk Score')
    
    plt.tight_layout()
    
    # Save frame to temp file
    temp_path = rf'C:\Projects\storm_risk\outputs\temp_frame_{i:03d}.png'
    plt.savefig(temp_path, dpi=80, bbox_inches='tight')
    temp_frames.append(temp_path)
    plt.close()
    
    if i % 10 == 0:
        print(f"  Frame {i+1} of {len(months)} done — {month}")

print("\nStitching frames into GIF...")

# Build GIF
with imageio.get_writer(gif_path, mode='I', duration=0.6, loop=0) as writer:
    for temp_path in temp_frames:
        image = imageio.imread(temp_path)
        writer.append_data(image)

# Clean up temp frames
for temp_path in temp_frames:
    os.remove(temp_path)

# Check file size
size_mb = os.path.getsize(gif_path) / (1024 * 1024)
print(f"\n✅ GIF exported successfully!")
print(f"Location: {gif_path}")
print(f"File size: {size_mb:.1f} MB")
print(f"Frames: {len(months)}")
```
